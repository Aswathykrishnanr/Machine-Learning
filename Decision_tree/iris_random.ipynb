```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Iris.csv')
df
```

```python
df.dtypes
```

```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
#from sklearn.ensemble import RandomForestClassifier  # or RandomForestRegressor

# Define model
#model = RandomForestClassifier(
#  n_estimators=100,        # number of trees
#  criterion='gini',        # or 'entropy', 'log_loss'
#  max_depth=None,          # tree depth (None = full depth)
#   max_features='sqrt',     # number of features per split
#   min_samples_split=2,     # min samples to split a node
#  min_samples_leaf=1,      # min samples in each leaf
# bootstrap=True,          # use bootstrap sampling
#   random_state=42,         # for reproducibility
#  n_jobs=-1                # use all CPU cores
# )
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,criterion='entropy',max_depth=10,max_features='sqrt')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
lab=['Iris-versicolor','Iris-virginica','Iris-setosa']
cmd=ConfusionMatrixDisplay(cm,display_labels=lab)
cmd.plot()
```
