```python
#get_dummies : after onehot encoding ,drop the first column then we get the values of each column
#get_dummies ===>encoding method in pandas 
```

```python
dfe = pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True,dtype=int)
dfe
```

```python
#concatenate df and dfe to get remaining columns
df_1 =pd.concat([df,dfe],axis=1)
df_1
```

```python
df_1.shape
```

```python
#drop
df_1.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'],axis=1,inplace=True)
df_1
```

```python
df_1.shape
```

```python
df_1.dtypes
```

```python
#string replace
df_1['Mileage']=df_1['Mileage'].str.replace('km/kg','')
df_1['Mileage']=df_1['Mileage'].str.replace('kmpl','')
df_1['Engine']=df_1['Engine'].str.replace('CC','')
df_1['Power']=df_1['Power'].str.replace('bhp','')
df_1
```

```python
#'null' value in some columns to replace with  0
df_1['Power'].unique()
```

```python
df_1['Mileage']=df_1['Mileage'].str.replace('null','0')
df_1['Engine']=df_1['Engine'].str.replace('null','0')
df_1['Power']=df_1['Power'].str.replace('null','0')
df_1.dtypes
```

```python
#data type change (string replace and splitting not changes the datatypes)
df_1['Engine'] = df_1['Engine'].astype(float)
df_1['Mileage'] = df_1['Mileage'].astype(float)
df_1['Power'] = df_1['Power'].astype(float)
df_1.dtypes
```

```python
#Mileage,Engine,Power ===> null value converted as '0'(obj)
#when dtype conversion ==>numerical value 0
#Replace 0 to NaN (Engine value cannot be 0)
#NaN represents missing value
df_1.loc[df_1.Engine==0,'Engine']=np.nan
df_1.loc[df_1.Power==0,'Power']=np.nan
df_1.loc[df_1.Mileage==0,'Mileage']=np.nan
```

```python
df_1.isna().sum()
```

```python
df_1['Engine']=df_1['Engine'].fillna(df_1['Engine'].mean())
df_1['Mileage']=df_1['Mileage'].fillna(df_1['Mileage'].mean())
df_1['Power']=df_1['Power'].fillna(df_1['Power'].mean())
df_1['Seats']=df_1['Seats'].fillna(df_1['Seats'].mode()[0])
df_1.isna().sum()
```

```python
x=df_1.drop(['Price'],axis=1)
x
```
